```python
df = pd.read_csv('Flights1_2019_1.csv')
df = df.drop(['Unnamed: 17'], axis=1)
df.columns
```

```python
import math
```

```python
def truncate_decimals(num, decimals):
    if math.isnan(num):
        return num
    factor = 10 ** decimals
    return int(num * factor) / factor
```

```python
list_ids = np.concatenate((df["ORIGIN_AIRPORT_ID"], df["DEST_AIRPORT_ID"]))
len(np.unique(list_ids))
```

```python
A_df = pd.read_csv('AIRPORT_COORDINATES.csv')
A_df['DISPLAY_AIRPORT_NAME'] = A_df['DISPLAY_AIRPORT_NAME'].str.replace("Airport", '')
A_df['DISPLAY_AIRPORT_NAME'] = A_df['DISPLAY_AIRPORT_NAME'].str.replace("-", ' ')
A_df['DISPLAY_AIRPORT_NAME'] = A_df['DISPLAY_AIRPORT_NAME'].str.replace("/", ' ')
A_df['LATITUDE']=  A_df['LATITUDE'].apply(truncate_decimals, decimals=2)
A_df['LONGITUDE']=  A_df['LONGITUDE'].apply(truncate_decimals, decimals=2)
#A_df['LATITUDE']= A_df['LATITUDE'].round(2)
#A_df['LONGITUDE'] = A_df['LONGITUDE'].round(2)
A_df = A_df.loc[A_df['ORIGIN_AIRPORT_ID'].isin(list_ids)].drop_duplicates()
A_df.head()
```

```python
B_df = pd.read_csv('airport_codes.csv')
B_df['Name'] = B_df['Name'].str.replace("Airport", '')
B_df['Name'] = B_df['Name'].str.replace("-", ' ')
B_df['Name'] = B_df['Name'].str.replace("/", ' ')
B_df['Latitude']=  B_df['Latitude'].apply(truncate_decimals, decimals=2)
B_df['Longitude']= B_df['Longitude'].apply(truncate_decimals, decimals=2)
#B_df['Latitude']= B_df['Latitude'].round(2)
#B_df['Longitude']= B_df['Longitude'].round(2)
B_df.head()
```

```python
new_df = pd.merge(A_df, B_df, how="inner", left_on=['LATITUDE','LONGITUDE'], right_on = ['Latitude','Longitude'])
new_df = new_df.drop([ 'City', 'Country'], axis=1)
new_df = new_df.drop_duplicates()

A_df['LATITUDE']=  A_df['LATITUDE'].apply(truncate_decimals, decimals=1)
A_df['LONGITUDE']=  A_df['LONGITUDE'].apply(truncate_decimals, decimals=1)
B_df['Latitude']=  B_df['Latitude'].apply(truncate_decimals, decimals=1)
B_df['Longitude']= B_df['Longitude'].apply(truncate_decimals, decimals=1)

new_df2 = pd.merge(A_df, B_df, how="left", left_on=['LATITUDE','LONGITUDE'], right_on = ['Latitude','Longitude'])
new_df2 = new_df2.loc[~new_df2['ORIGIN_AIRPORT_ID'].isin(new_df["ORIGIN_AIRPORT_ID"])].drop_duplicates().dropna()

new_df = pd.concat([new_df, new_df2])
```

```python
new_df = new_df.drop([ 'City', 'Country', 'DISPLAY_AIRPORT_NAME', "LATITUDE", "LONGITUDE"], axis=1)
new_df
```
