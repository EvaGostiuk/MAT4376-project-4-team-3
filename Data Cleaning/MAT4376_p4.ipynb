```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import seaborn as sns
import matplotlib.pylab as plt
import plotly.express as px

import copy

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Flights1_2019_1.csv')
df = df.drop(['Unnamed: 17'], axis=1)
df.columns
```

```python
df['ORIGIN_STATE_ABR'] = df['ORIGIN_CITY_NAME'].str[-2:]

df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 1, 'Mon', df['DAY_OF_WEEK'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 2, 'Tue', df['DAY_OF_WEEK_STR'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 3, 'Wed', df['DAY_OF_WEEK_STR'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 4, 'Thu', df['DAY_OF_WEEK_STR'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 5, 'Fri', df['DAY_OF_WEEK_STR'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 6, 'Sat', df['DAY_OF_WEEK_STR'])
df['DAY_OF_WEEK_STR'] = np.where(df['DAY_OF_WEEK'] == 7, 'Sun', df['DAY_OF_WEEK_STR'])

df['DEP_DELAY_NEW'] = np.where(df['DEP_DELAY'] <= 0, 0, df['DEP_DELAY'])
df['DEP_DEL15'] = np.where(df['DEP_DELAY'] >= 15, 1, 0)
df['DEP_DEL15'] = np.where(df['DEP_DELAY'] .isna(), np.NAN, df['DEP_DELAY'])
```
